```python
#importowanie biblioteki
import matplotlib.pyplot as plt
```

```python
# Dane do wykresu
data = [25, 35, 20, 20]

plt.pie(data)
```

```python
labels = ['Jabłka', 'Banany', 'Pomarańcze', 'Winogrona']
plt.pie(data, labels=labels)
```

```python
#Dodanie tytułu - metoda title
plt.pie(data, labels=labels)
plt.title("Owoce")
```

```python
#Zmiena kolorów - parametr colors
colors = ['red','blue','green','yellow']
plt.pie(data, labels=labels, colors=colors)
plt.title("Owoce")
```

```python
# dodac kolory w palecie
```

```python
#dodanie wartości procentowych na wykresie - autopct
plt.pie(data, labels=labels, autopct='%.1f%%')
plt.title("Owoce")
```

```python
#dodać wykres z trudniejszymi danymi, pokazac inne opcje zaokrąglenia
```

```python
#oddalenie wartości procentowych od środka - pctdistance
plt.pie(data, labels=labels, autopct='%.1f%%', pctdistance=0.8)
plt.title("Owoce")
```

```python
#dodanie cienia do wykresu
plt.pie(data, labels=labels, autopct='%.1f%%', pctdistance=0.8, shadow=True)
plt.title("Owoce")
```

```python
# poszukać czy da się obrócić wykres
```

```python
#Odległość etykiet od środka wykresu
plt.pie(data, labels=labels, autopct='%.1f%%', labeldistance=1.5)
plt.title("Owoce")
```

```python
#ustawienie kątu od którego zaczyna się rysować wykres
plt.pie(data, labels=labels, autopct='%.1f%%', startangle=90)
plt.title("Owoce")
```

```python
#zmiana wielkości wykresu - radius
plt.pie(data, labels=labels, autopct='%.1f%%', radius=2)
plt.title("Owoce")
```

```python
#ustawia kierunek rysowania wykresu
plt.pie(data, labels=labels, autopct='%.1f%%', counterclock=False)
plt.title("Owoce")
```

```python
#ustawia kierunek etykiet względem wykresu
plt.pie(data, labels=labels, autopct='%.1f%%', rotatelabels=True)
plt.title("Owoce")
```

```python
#odsuwa od środka wykresu jego części
explode=[0,0.1,0,0]
plt.pie(data, explode=explode, labels=labels, autopct='%.1f%%')
plt.title("Owoce")
```

```python
wedges = plt.pie(data, explode=explode, labels=labels, autopct='%.1f%%')[0]

# Zmiana właściwości tylko drugiego klina (indeks 1)
wedges[1].set_edgecolor('green')
wedges[1].set_linewidth(3)
wedges[1].set_linestyle('dotted')
```

```python
wedges = plt.pie(data, explode=explode, labels=labels, autopct='%.1f%%', textprops={'fontsize': 12, 'color': 'blue'})[0]

# Zmiana właściwości tylko drugiego klina (indeks 1)
wedges[1].set_edgecolor('green')
wedges[1].set_linewidth(3)
wedges[1].set_linestyle('dotted')
```

```python
wedges, texts, autotexts = plt.pie(data, explode=explode, labels=labels, autopct='%.1f%%')

# Zmiana właściwości tylko drugiego klina (indeks 1)
wedges[1].set_edgecolor('green')
wedges[1].set_linewidth(3)
wedges[1].set_linestyle('dotted')

texts[1].set_fontsize(14)  # Zmieniamy wielkość czcionki dla etykiety
texts[1].set_color('red')  # Zmieniamy kolor czcionki dla etykiety

autotexts[1].set_fontsize(14)  # Zmieniamy wielkość czcionki dla wartości procentowej
autotexts[1].set_color('blue')  # Zmieniamy kolor czcionki dla wartości procentowej
```
